```python
# Parsing text into words
text = [ 'All that we are is the result of what we have thought' , 'To be or not to be that is the question' , 'Be yourself everyone else is already taken' ]
text
```

```python
import re
re.split('\\s',text[0])
```

```python
' '.join(re.split('\\s',text[0]))
```

```python
allwords = re.split('\\s' , ' '.join(text).lower() )
allwords
```

```python
# find the unqiue set of words
vocab = sorted(set(allwords))
vocab
```

```python
print(f"There are {len(allwords)} in the text and {len(vocab)} unique words")
```

```python
# create encoder
word2idx = {}
for i,word in enumerate(vocab):
  word2idx[word] = i
word2idx
```

```python
# and decoder
idx2word = {}
for i , word in enumerate(vocab):
  idx2word[i] = word
idx2word
```

```python
import numpy as np
randidx = np.random.randint(0,len(vocab),5)
# [idx2word[i] for i in randidx]
' '.join([idx2word[i] for i in randidx])
```

```python
# tokenization
text_as_int = [word2idx[word] for word in allwords]
text_as_int
```

```python
for tokeni in text_as_int:
  print(f"Token {tokeni:2} : {idx2word[tokeni]}")
```
