```python
# Parsing text into words
text = [ 'All that we are is the result of what we have thought' , 'To be or not to be that is the question' , 'Be yourself everyone else is already taken' ]
text
```

```python
import re

allwords = re.split('\\s', ''.join(text).lower() )
vocab = set(allwords)
vocab
```

```python
word2idx = {word:i for i,word in enumerate(vocab)}
word2idx
```
